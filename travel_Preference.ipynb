### Data Preparations 
Here we will do encoding and prepare for the advance exploratory analysis

### Phase 3 : K means clustering 
Unsupervised clustering was performed before supervised classification to understand inherent data structure prior to label-based prediction.
To identify natural groupings of travelers based on their preferences and behavior, without using the target travel preference.

```python
#exporting the dataset after preparation
df.to_csv("ml_ready_travel_preference_data.csv", index=False)
```

```python
#defining ideal column for use
cluster_features = [
    'likes_water',
    'likes_altitude',
    'adventure_level',
    'travel_frequency',
    'trip_duration',
    'budget'
]

X_cluster = df[cluster_features]
```

```python
#elbow method of selecting k
from sklearn.cluster import KMeans

wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_cluster)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.show()
```

```python
#assunming k=3 as clusters
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_cluster)

df['cluster'] = clusters
```

```python
#analysing our cluster
df['cluster'].value_counts()
```

```python
#cluster vs actual preference
pd.crosstab(df['cluster'], df['target'])
```

```python
#cluster visualization 
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_cluster)

df['PC1'] = X_pca[:, 0]
df['PC2'] = X_pca[:, 1]

sns.scatterplot(
    x='PC1',
    y='PC2',
    hue='cluster',
    data=df,
    palette='Set2'
)
plt.title("K-Means Clusters Visualized using PCA")
plt.show()
```

### Phase 4 Logistic Regression 
To predict a person’s preferred travel destination (Hills / Plains / Beaches) using behavioral and demographic features.

```python
from sklearn.model_selection import train_test_split

#columns seelection
X = df.drop(columns=['target', 'activities', 'cluster', 'PC1', 'PC2'], errors='ignore')
y = df['target']

#splitting in train and test using stratified 
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

```python
#train logistic regresssion(multiclass)
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(
    multi_class='ovr',
    max_iter=1000,
    class_weight='balanced'
)
log_reg.fit(X_train, y_train)
```

```python
#making prediction
y_pred = log_reg.predict(X_test)
y_prob = log_reg.predict_proba(X_test)
```

```python
#model evaluation
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

While clustering revealed partial natural grouping, logistic regression significantly improves predictive accuracy by learning class boundaries from labeled data. 
Made the prediction with accuracy overall of 64%.

## Phase 5 KNN
For better prediction with smaller model. 
Train test split will be resued from logistical regression                                                     

```python
#we have already made train test split and will use it 
#we are now choosing K 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score

k_values = range(3, 16, 2)
f1_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    f1 = f1_score(y_test, y_pred, average='macro')
    f1_scores.append(f1)

plt.plot(k_values, f1_scores, marker='o')
plt.xlabel("K value")
plt.ylabel("Macro F1-score")
plt.title("Choosing Optimal K for KNN")
plt.show()
```

```python
#assuming  k = 9 with the highest Macro F1 Score
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
```

```python
#confusion matrix and others are already imported in logistic

print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print("\nClassification Report:\n", classification_report(y_test, y_pred_knn))

cm = confusion_matrix(y_test, y_pred_knn)
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
plt.title("Confusion Matrix - KNN")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

#### observation
Despite stratified splitting and balanced class weights, the Plains category remained difficult to predict due to limited samples and overlapping preference patterns. This shows challenge of minority class classification in real-world survey data.

### Phase 8 Apriori 
To discover frequent activity patterns and their association with travel destination preferences.

```python
# Make a copy for Apriori
apriori_df = df[['activities', 'target']].copy()

# Clean text
apriori_df['activities'] = apriori_df['activities'].str.lower().str.strip()

# Split into lists
apriori_df['activities'] = apriori_df['activities'].str.split(',')
apriori_df['activities']
```

```python
#one hot encoding for apriori using transactionencoder instead of scikit labelencoder
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

te = TransactionEncoder()
te_array = te.fit(apriori_df['activities']).transform(apriori_df['activities'])

activity_df = pd.DataFrame(te_array, columns=te.columns_)
activity_df.head()

#applying apriori model on our dataset
frequent_itemsets = apriori(
    activity_df,
    min_support=0.05,
    use_colnames=True
)
frequent_itemsets.sort_values('support', ascending=False).head(10)
```

```python
#generating association rules
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.6
)

rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].sort_values(by='lift', ascending=False).head(10)
```

```python
apriori_df['destination'] = apriori_df['target'].map({
    0: 'hills',
    1: 'plains',
    2: 'beaches'
})

transactions = apriori_df['activities'].tolist()

# Append destination
transactions = [
    acts + [dest]
    for acts, dest in zip(apriori_df['activities'], apriori_df['destination'])
]
```

```python
#encoding again 
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
trans_df = pd.DataFrame(te_array, columns=te.columns_)
```

```python
#running apriori 
frequent_itemsets = apriori(
    trans_df,
    min_support=0.05,
    use_colnames=True
)

rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.6
)

rules[rules['consequents'].astype(str).str.contains('hills|beaches|plains')].sort_values(by='lift', ascending=False).head(10)
```
